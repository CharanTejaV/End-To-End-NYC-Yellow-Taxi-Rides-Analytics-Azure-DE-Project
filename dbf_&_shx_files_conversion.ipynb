```python
from dbfread import DBF
import csv

# Function to convert DBF to CSV
def convert_dbf_to_csv(dbf_file, csv_file):
    table = DBF(dbf_file)
    with open(csv_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(table.field_names)
        for record in table:
            writer.writerow(list(record.values()))

# Replace 'your_file.dbf' with the path to your DBF file
# Replace 'output_file.csv' with the desired name for the CSV file
convert_dbf_to_csv('C:/Users/princ/Downloads/taxi_zones.dbf', 'C:/Users/princ/Downloads/taxi_zones.csv')
```

```python
import geopandas as gpd

# Replace 'path/to/taxi_zones.shp' with the actual path to your shapefile
shapefile_path = 'C:/Users/princ/Downloads/taxi_zones/taxi_zones.shp'

# Read the shapefile
gdf = gpd.read_file(shapefile_path)

gdf = gdf.to_crs(epsg='4326')

# List to store latitude and longitude for each zone
zone_coordinates = []

# Iterate through each row in the DataFrame
for index, row in gdf.iterrows():
    # Access the geometry for the current row
    geometry = row['geometry']
    
    # Access latitude and longitude for the centroid of the polygon
    latitude, longitude = geometry.centroid.y, geometry.centroid.x
    
    # Append the coordinates to the list
    zone_coordinates.append((latitude, longitude))

# Print the list of coordinates
print(zone_coordinates)
```

```python
import geopandas as gpd
import pandas as pd

# Replace 'path/to/taxi_zones.shp' with the actual path to your shapefile
shapefile_path = 'C:/Users/princ/Downloads/taxi_zones/taxi_zones.shp'

# Read the shapefile
gdf = gpd.read_file(shapefile_path)

# Reproject to WGS84 (latitude and longitude)
gdf = gdf.to_crs(epsg='4326')

# Create a new DataFrame to store the results
result_df = pd.DataFrame(columns=['Zone', 'LocationID', 'Latitude', 'Longitude'])

# Iterate through each row in the DataFrame
for index, row in gdf.iterrows():
    # Access the attributes and geometry for the current row
    zone = row['zone']
    location_id = row['LocationID']
    geometry = row['geometry']
    
    # Access latitude and longitude for the centroid of the polygon
    latitude, longitude = geometry.centroid.y, geometry.centroid.x
    
    # Append the results to the new DataFrame
    #result_df = result_df.append({'Zone': zone, 'LocationID': location_id, 'Latitude': latitude, 'Longitude': longitude}, ignore_index=True)
    result_df = pd.concat([result_df, pd.DataFrame({'Zone': [zone], 'LocationID': [location_id], 'Latitude': [latitude], 'Longitude': [longitude]})], ignore_index=True)

result_df.to_csv('C:/Users/princ/Downloads/taxi_zones/zone_coordinates.csv', index=False)
```
